#### a) Coluna Atrito (Taxa de desgaste entre os membros da equipe)

```python
# Verificando qual  a taxa de desgate na empresa

# Quantidade de valor sim e no = df_funcionario.Atrito.value_counts() >>> Resposta: No = 1233 e Sim = 237
# Total de funcionarios = len(df_funcionario) >>> Resposta: 1470
# % = (1233/1470) x 100 e (237/1470) x 100
df_funcionario.Atrito.value_counts()/len(df_funcionario)*100

# Agrupando pela coluna Atrito e realizando a contagem da quantidade dos seus valores  
# Mostra a quantidade de funcionrio agrupados pela existencia de atrito ou no
df_atrito = df_funcionario.groupby('Atrito')['Atrito'].count()

# Grafico de pizza ilustrando o agrupamento da coluna Atrito com base na contagem dos seus valores
plt.pie(df_atrito, explode=[0.05,0.1], labels=['No', 'Sim'], radius=1.0, autopct='%1.1f%%', shadow=False);
# explode = Destacando a fatia menor
# labels = Legenda de cada fatia
# radius = Valor do raio 
# autopct = formato da porcentagem
# shadow = Sem sombra no grafico
```

#### Anlise: A coluna Atrito  composta por dois valores: "NO", indicando que o funcionrio no experimentou nenhum tipo de desgaste na empresa, e "SIM", representando o caso contrrio, em que o funcionrio teve algum tipo de desgaste. De acordo com o clculo realizado no perodo de 03/01/2012 a 31/12/2022, foi constatado um ndice de atrito de 16,12%, correspondendo a um total de 1.470 funcionrios.

#### b) Criao da funo para transformar os valores da coluna Atrito (Sim = 1 e No = 0) e aplicao na coluna Atrito

```python
# Se o valor for igual a Sim retorna 1, caso o valor seja igual a No retorne 0
def sim_nao_para_1_0(valor):
    if valor == 'Sim':
        return 1
    elif valor == 'No':
        return 0
    else:
        return valor
    
# Aplicando a funo sim_nao_para_1_0 na coluna de Atrito e seu armazenamento em um novo dataframe
df_funcionario['Atrito'] = df_funcionario['Atrito'].map(sim_nao_para_1_0)
df_funcionario['Atrito'].head()
# A funo map est aplicando a funo sim_nao_para_1_0 criada em todas as linhas da coluna Atrito. 
```

#### c) Coluna de gnero

```python
# Agrupando pela coluna Gnero e realizando a contagem de seus valores  
# Mostra a quantidade de funcionrio agrupados por gnero
df_genero = df_funcionario.groupby('Gnero')['Gnero'].count()
display(df_genero)

# Visualizando a proporo de gnero atravs do grfico de contagem
sns.countplot(data = df_funcionario, x = 'Gnero')
plt.title('Proporo de gnero');
```

```python
# Agrupando pelas colunas Gnero e Atrito e realizando a contagem da quantidade de Gnero. Exibindo como tabela (to_frame)
df_funcionario.groupby(['Gnero', 'Atrito'])['Gnero'].count().to_frame() 
```

```python
# Agrupando pela coluna Gnero e realizando a mdia dos valores da coluna atritos. Exibindo como tabela (to_frame)
# sort_values('Atrito', ascending=False) = ordenando pelos valores da coluna de atrito na ordem decrescente
# Usar a mdia em vez da contagem remove o ruiod e fornece informaes mais claras sobre os dados com base na proporo de gnero.
df_funcionario.groupby('Gnero')['Atrito'].mean().to_frame().sort_values('Atrito', ascending=False)  
```

```python
# Criando uma coluna de Renda mensal do salrio
df_funcionario['Renda mensal'] = df_funcionario['Salrio'] / 12

# Agrupando pelas colunas Gnero e Atrito e realizando a mdia da Renda Mensal. Exibindo como tabela (to_frame)
df_funcionario.groupby(['Gnero', 'Atrito'])['Renda mensal'].mean().to_frame() 
```

#### Anlise: De acordo com os clculos realizados no perodo de 03/01/2012 a 31/12/2022, foi constatado que os funcionrios do gnero Masculino esto deixando a empresa mais do que os funcionrios do gnero feminino.
```python
# Criao da funo nivel_educacao para conectar os valores do Dataframe de df_funcionario e df_nivel_educacional

def nivel_educacao(ID_nivel):
    if ID_nivel == 1:
        return 'Sem qualificaes formais'
    elif ID_nivel == 2:
        return 'Ensino mdio'
    elif ID_nivel == 3:
        return 'Bacharis'
    elif ID_nivel == 4:
        return 'Mestres'
    else:
        return 'Doutorado'
    
# Criao da Dataframe Educacao usando o valor da coluna Educao e aplicando a funo nivel_educacao
df_educacao = df_funcionario['Educao'].map(nivel_educacao)

# Criao da coluna Nova educao utilizando o Dataframe Educacao
df_funcionario['Nova_educao'] = df_educacao

# Mostra os valores da coluna Educao com os valores do Dataframe df_nivel_educacional 
display(df_funcionario['Nova_educao'].head(3))
```

```python
# Agrupando pelas colunas Nova_educao e Atrito e realizando a mdia pela quantidade de atrito
# sort_values('Atrito', ascending=False) = ordenando pelo valor da coluna de Atrito na ordem decrescente
df_funcionario.groupby('Nova_educao')['Atrito'].mean().to_frame().sort_values('Atrito', ascending=False) # to_frame = Mostrando com tabela
```

```python
# Agrupando pelas colunas Nova_educao e Atrito e realizando a contagem pela quantidade de atrito
df_funcionario.groupby(['Nova_educao', 'Atrito'])['Atrito'].count().to_frame() # to_frame = Mostrando com tabela
```

```python
# Agrupando pela coluna Nova_educao e realizando a mdia da Renda mensal
# sort_values('Renda mensal') = ordenando pelo valor da coluna da Renda mensal na ordem crescente
df_funcionario.groupby('Nova_educao')['Renda mensal'].mean().to_frame().sort_values('Renda mensal') # to_frame = Mostrando com tabela
```
