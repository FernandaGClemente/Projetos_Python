### Importação das Bibliotecas

### Leitura e armazenamento dos dados do CSV num DataFrame

```python
df_exames = pd.read_csv('_datasets/StudentsPerformance.csv')
display(df_exames)
```

### Formas de exibição do DataFrame

```python
df_exames.head()
# df_exames.tail()
# pd.set_option('display.max_rows', 1000)
# df_exames
```

### Métodos do DataFrame

```python
# Mostra a quantidade de linhas e colunas
df_exames.shape
# Mostra o número do indice inicial, o número do indice final e de quanto em quanto está sendo incrementado 
df_exames.index
# Exibição dos atributos (Colunas)
df_exames.columns
# Mostra o tipo de dado de cada coluna
df_exames.dtypes
# Mostra as informações dos dados contidos no Dataframe
df_exames.info()
# Mostra algumas estatisticas descritivas das colunas númericas
df_exames.describe()
```

### Funções úteis

```python
# Obtendo o comprimento do dataframe (número de linhas)
len(df_exames)
# Obtendo o índice mais alto do dataframe
max(df_exames.index)
# Obtendo o índice mais baixo do dataframe
min(df_exames.index)
# Obtendo o tipo de dados
type(df_exames)
# Arredondando os valores do conjunto de dados para duas casas decimais
round(df_exames, 2)
```

### Analisando apenas uma coluna

```python
# Formas de selecionar apenas uma coluna 
df_exames['gender']
df_exames.gender
df_exames[['gender']]
# Mostra o tipo de dados de uma coluna
type(df_exames['gender'])
# Pode utilizar os métodos e as funções na série / coluna
df_exames['gender'].index
df_exames['gender'].head()
```

### Analisando mais de uma coluna

```python
# Formas de selecionar duas ou mais colunas
df_exames[['gender', 'math score']]
df_exames[['gender', 'math score', 'reading score', 'writing score']]
# Mostra o tipo de dados das colunas
type(df_exames[['gender', 'math score']])
```

### Formas de adicionar uma nova coluna

```python
# Adicionando a coluna "Pontuação de idioma" onde todas as linhas receberá o número 70
df_exames['language score'] = 70
# Adicionando a coluna "Pontuação de idioma" onde todas as linhas receberá os números 0 a 1000
language_score = np.arange(0, 1000)
df_exames['language score'] = language_score
# Adicionando a coluna "Pontuação de idioma" onde todas as linhas receberá números inteiros aleatorios de 1 a 100 
int_language_score = np.random.randint(1, 100, size=1000)
df_exames['language score'] = int_language_score
df_exames
```

```python
# criar números flutuantes aleatórios entre 1 e 100
np.random.uniform(1, 100, size=1000)
```

### Operações matemáticas em colunas

```python
# Soma de todos os valores da coluna "Pontuação de matemática"
df_exames['math score'].sum()
# Conta a quantidade de valores da coluna "Pontuação de matemática"
df_exames['math score'].count()
# Calcula a média dos valores da coluna "Pontuação de matemática"
df_exames['math score'].mean()
# Calcula o desvio padrão dos valores da coluna "Pontuação de matemática"
df_exames['math score'].std()
# Mostra o valor mais alto da coluna "Pontuação de matemática"
df_exames['math score'].max()
# Mostra o valor mais baixo da coluna "Pontuação de matemática"
df_exames['math score'].min()
```

### Operações matemáticas em linhas

```python
# Soma de todos os valores das linhas das colunas "Pontuação de matemática", "Pontuação de leitura" e "Partitura escrita"
df_exames['math score'] + df_exames['reading score'] + df_exames['writing score']
# Calcula a média dos valores das linhas das coluna "Pontuação de matemática", "Pontuação de leitura" e "Partitura escrita"
df_exames['average'] = (df_exames['math score'] + df_exames['reading score'] + df_exames['writing score'])/3
df_exames['average']
```

### Formas de realizar contagens de valores

```python
# Contando a quantiodade de elementos da coluna gênero
len(df_exames['gender'])
df_exames['gender'].count()
# Contando a quantidade de elementos da coluna gênero por categoria
df_exames['gender'].value_counts()
# Retorna a frequência relativa (divide todos os valores pela soma dos valores) da coluna gênero por categoria
df_exames['gender'].value_counts(normalize=True)
# Retornar a frequência relativa e arredondar para 2 casas decimais da coluna "nível de escolaridade dos pais" por categoria
df_exames['parental level of education'].value_counts(normalize=True).round(2)
```

### Formas de realizar classificação dos valores

```python
# Classificando o DataFrame utilizando os dados da coluna "Pontuação de matemática" como base (Ordem Crescente)
df_exames.sort_values('math score')
# Classificando o DataFrame utilizando os dados da coluna "Pontuação de matemática" como base (Ordem Decrescente)
df_exames.sort_values('math score', ascending=False)
# Classificando o DataFrame utilizando os dados da coluna "Pontuação de matemática" 
# e depois da coluna "Pontuação de leitura" como base (Ordem Decrescente)
df_exames.sort_values(['math score', 'reading score'], ascending=False)
# Classificando o DataFrame utilizando os dados da coluna "Pontuação de matemática" 
# e depois da coluna "Pontuação de leitura" como base (Ordem Decrescente) e atualizando o DataFrame
df_exames.sort_values(['math score', 'reading score'], ascending=False, inplace=True)

# Classificando o DataFrame na ordem decrescente da coluna "Raça/etnia" utilizando a função chave (key), pois se trata de uma coluna com texto e não números
# A função lambda, contida na chave, está colocando todos os dados da coluna "Raça/etnia" com letras minusculas, 
# por que só assim será possivel fazer a classificação do texto 
df_exames.sort_values(['race/ethnicity'], ascending=True, key=lambda a:a.str.lower())
```

### Criação de tabelas dinâmicas

```python
df_gdp = pd.read_csv('_datasets/gdp.csv', encoding='Latin 1')
df_gdp
```

```python
# A coluna "Ano" é colocado como indice, agrupará os dados das linhas
# Os dados da coluna "Cidade" que estavam na linha passaram a ser os dados das colunas
# Os dados da coluna "PIB per capita" são colocados nas linhas de acordo com o ano e a cidade
df_gdp.pivot(index="year", columns="country", values="gdppc")
```

```python
df_sales = pd.read_excel('_datasets/supermarket_sales.xlsx')
df_sales
```

```python
# A coluna "Genero" é colocado como indice, agrupará os dados das linhas
# A função de agregação Soma totalizará os valores de todas as colunas numéricas agrupando por gênero
df_sales.pivot_table(index='Gender',
                     aggfunc='sum').round(1)
```

```python
# A coluna "Genero" é colocado como indice, agrupará os dados das linhas
# A função de agregação Soma totalizará os valores das colunas "Quantidade" e "Total" agrupando por gênero
df_sales.pivot_table(index='Gender',
                     values=['Quantity', 'Total'],
                     aggfunc='sum').round(1)
```

```python
# A coluna "Genero" é colocado como indice, agrupará os dados das linhas
# A função de agregação Soma totalizará os valores da coluna "Total" para cada coluna dos dados da coluna "Linha de produto" agrupando por gênero
df_sales.pivot_table(index='Gender',
                     columns='Product line',
                     values='Total',
                     aggfunc='sum').round(1)
```

```python
df_population_raw = pd.read_csv('_datasets/population_total.csv')
df_population_raw
```

```python
# Descartando valores nulos
df_population_raw.dropna(inplace=True)
```

```python
# A coluna "Ano" é colocado como indice, agrupará os dados das linhas
# Os dados da coluna "Cidade" que estavam na linha passaram a ser os dados das colunas
# Os dados da coluna "população" são colocados nas linhas de acordo com o ano e a cidade
df_pivot = df_population_raw.pivot(index='year', columns='country',values='population')
df_pivot
```

```python
# Selecionando alguns países
df_pivot = df_pivot[['United States', 'India', 'China', 'Indonesia', 'Brazil']]
df_pivot
```

### Gráfico de Linha

```python
# Kind = Tipo de grafico, como é um grafico de linha colocamos Line
# xlabel = legenda do eixo x
# ylabel = Legenda do eixo y
# Title = Título do gráfico
# figsize = Toamanho da figura branca que serve como tela de fundo para o gráfico
df_pivot.plot(kind='line', xlabel='Years', ylabel='Population', title='Population (1955-2020)',figsize=(8,4));
```

### Gráfico de Barra com apenas uma variável

```python
# Selecionando apenas um ano (2020)
df_pivot_2020 = df_pivot[df_pivot.index.isin([2020])]
df_pivot_2020
```

```python
# Transformando as linhas em colunas 
df_pivot_2020 = df_pivot_2020.T
df_pivot_2020
```

```python
# Kind = Tipo de grafico, como é um grafico de barra colocamos bar
# color = cor das barras
# xlabel = legenda do eixo x
# ylabel = Legenda do eixo y
# Title = Título do gráfico
df_pivot_2020.plot(kind='bar', color='orange', xlabel='Years', ylabel='Population', title='Population in 2020');
```

### Gráfico de Barras agrupado por “n” variáveis

```python
# Selecionando alguns anos
df_pivot_sample = df_pivot[df_pivot.index.isin([1980, 1990, 2000, 2010, 2020])]
df_pivot_sample
```

```python
# Kind = Tipo de grafico, como é um grafico de barra colocamos bar
# xlabel = legenda do eixo x
# ylabel = Legenda do eixo y
# Title = Título do gráfico
# figsize = Toamanho da figura branca que serve como tela de fundo para o gráfico
df_pivot_sample.plot(kind='bar', xlabel='Years', ylabel='Population', figsize=(8,4));
```

### Gráfico de Pizza

```python
# Alterando o nome da coluna "Ano" de 2020.0 para 2020
df_pivot_2020.rename(columns={2020:'2020'}, inplace=True)
df_pivot_2020
```

```python
# Kind = Tipo de grafico, como é um grafico de pizza colocamos pie
# y = Informamos qual será os dados do eixo y
# Title = Título do gráfico
df_pivot_2020.plot(kind='pie', y='2020', title='Population in 2020 (%)');
```
