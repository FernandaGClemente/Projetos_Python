# **Códigos - Divisão de dados (Data Splitting)**

<br>

#### Este tutorial aborda como dividir dados em conjuntos de dados de teste (Por padrão: 20%) e treinamento (Por padrão: 80%).
#### Sem a divisão, os dados podem sofrer overfitting (Superajuste dos dados da população) ou underfitting (Subajuste dos dados da população) interferindo na análise. 

<br>

## Outra forma de realizar a divisão dos dados:
#### 1 - Dado originais = 100%
#### 2 - Exploração de dados = 80% e Dados para validação (Teste 1) = 20%
#### 3 - Exploração de dados >>> Dados de treinamentos = 80% e Dados para teste (Teste 2) = 20%

#### 1 - Importação dos dados (Conjunto de casas para alugar em diferentes cidades do Brasil)
> https://www.kaggle.com/datasets/rubenssjr/brasilian-houses-to-rent?resource=download

```python
import pandas as pd

habitacao = pd.read_csv("houses_to_rent.csv")
display(habitacao)
```

#### 2 - Crie o dataframe x com variáveis preditoras e o dataframe y com variável de resultado.

```python
# X = Colunas = Recursos (Feature) = variáveis preditorias
# Y = Coluna de resultado = variavel de resultado
# linhas representa as amostras (observações)
df_X = habitacao.iloc[:,1:13]
df_X
```

```python
df_Y = habitacao.loc[:,"total"]
df_Y
```

#### 3 - Realizando a divisão do conjunto de dados em Teste e em Treinamento

```python
from sklearn.model_selection import train_test_split

# test_size=.2 é o tamanho da amostra do conjunto teste = 20% = 0.2 = .2
train_test_split(df_X,df_Y,test_size=.2)
```

#### 4 - Salvando as informações nas determinadas variáveis

```python
X_train, X_test, y_train, y_test = train_test_split(df_X,df_Y,test_size=.2)
```

#### Adicione um parâmetro opcional para definir uma semente. <br>
##### random_state defini a semente para garantir que obterá a mesma divisão todas as vezes, ou seja, 
##### Ter certeza que não está dividindo aleatóriamente, mas que está dividindo da mesma maneira todas as vezes
##### 1337 é um número qualquer de semente 

```python
X_train, X_test, y_train, y_test = train_test_split(df_X,df_Y,test_size=.2, random_state = 1337)
```
